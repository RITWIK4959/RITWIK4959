```python
df.replace({'Fuel':{'Diesel':0 ,'Petrol':1,'CNG':2,'LPG':3,'Electric':4}},inplace=True)
```

```python
df
```

```python
df.replace({'Seller_Type':{'Individual':0,'Dealer':1,'Trustmark Dealer':2}},inplace=True)
```

```python
df.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
df.replace({'Owner':{'First Owner':0,'Second Owner':1,'Third Owner':2,'Fourth & Above Owner':3,'Test Drive Car':4}},inplace=True)
```

```python
df.replace({'Brand':{'Maruti':0,'Hyundai':1,'Mahindra':2,'Honda':3,'Tata':4,'Ford':5,'Toyota':6,'Chevrolet':7,'Renault':8,'Volkswagen':9,'Skoda':10,'Nissan':11,'Audi':12,'BMW'	:13,'Fiat':14,'Datsun':15,'Mercedes-Benz':16,'Mitsubishi':17,'Jaguar':18,'Land':19,'Ambassador':20,'Volvo':21,'Jeep':22,'OpelCorsa':23,'MG':24,'Isuzu':25,'Force':26,'Daewoo':27,'Kia':28}},inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['Model']=le.fit_transform(df['Model'])
```

```python
df['Model']
```

```python
y=df['Selling_Price']
x=df[['Brand','Model','Year','KM_Driven', 'Fuel','Seller_Type', 'Transmission', 'Owner']]
```

```python
y.shape
```

```python
df['Brand']
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2456)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
```

```python
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel("Actual")
plt.ylabel("predicted")
plt.title("Actual vs predicted")
plt.show()
```
